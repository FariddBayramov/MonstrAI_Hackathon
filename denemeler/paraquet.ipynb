```python
import dask.dataframe as dd

# Dask ile Parquet dosyasını oku
df = dd.read_parquet('data/MobilityDataMay2024.paraquet')

# Veri tiplerini göster
print("\nData Types:")
print(df.dtypes)

# Bölüm sayısını göster
print("\nNumber of Partitions:")
print(df.npartitions)

num_rows = df.shape[0].compute() # Satır sayısı Dask Future'ı olduğu için .compute() çağırıyoruz.
num_cols = df.shape[1]         # Sütun sayısı zaten bir int olduğu için doğrudan erişiyoruz.

print("\nShape (Rows, Columns):")
print(f"({num_rows}, {num_cols})")
```
