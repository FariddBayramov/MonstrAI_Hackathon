```python
# Define groups based on user instructions
b_group = ['NAMIK KEMAL', 'ESENEVLER', 'IHLAMURKUYU']  # Adjust 'ESENEVLER' if needed
a_group = [
    'ATATÜRK', 'BARBAROS', 'İÇERENKÖY', 'MERKEZ', 'YENİŞEHİR', '19 MAYIS',
    # All neighborhoods from ACIBADEM through ZÜHTÜPAŞA inclusive
    'ACIBADEM', 'BOSTANCI', 'CADDEBOSTAN', 'CAFERAĞA', 'DUMLUPINAR', 'EĞİTİM',
    'FENERBAHÇE', 'FENERYOLU', 'FİKİRTEPE', 'GÖZTEPE', 'HASANPAŞA', 'KOŞUYOLU',
    'KOZYATAĞI', 'MERDİVENKÖY', 'MERKEZ-ERENKÖY', 'OSMANAĞA', 'RASİMPAŞA',
    'SAHRAYCEDİD', 'SUADİYE', 'ŞAŞKINBAKKAL', 'ZÜHTÜPAŞA'
]
c_group = [
    'ESAT PAŞA', 'İNÖNÜ', 'KAYIŞDAĞI', 'KÜÇÜKBAKKALKÖY', 'ÖRNEK',
    'ALTINŞEHİR', # neighborhoods before AHMEDİYE not counted in A or B
]

# Anything after AHMEDİYE (inclusive) is A
after_ahmediye = df['Mahalle'] >= 'AHMEDİYE'  # example condition, adjust as needed

# Create the 'Mahalle Grubu' column
conditions = [
    df['Mahalle'].isin(b_group),
    df['Mahalle'].isin(a_group),
    df['Mahalle'].isin(c_group)
]
choices = ['B', 'A', 'C']

df['Mahalle Grubu'] = np.select(conditions, choices, default='A')  # default to A for any remaining
```

```python
# LabelEncoding "Mahalle Grubu"
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Mahalle Grubu'] = le.fit_transform(df['Mahalle Grubu'])
df
```

```python
df = pd.get_dummies(df, columns=["Mahalle"], prefix=["Mahalle"], drop_first=True, dtype=float)
df
```

```python
df = pd.get_dummies(df, columns=["İlçe"], prefix=["İlçe"], drop_first=True, dtype=float)
df
```

```python
df = pd.get_dummies(df, columns=["Tip"], prefix=["Tip"], drop_first=True, dtype=float)
df
```

```python
df.drop(columns=["Mahalle Grubu"], inplace=True)
```

```python
df.columns
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
inertia = []
silhouette = []

K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(df, kmeans.labels_))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(K_range, inertia, marker='o')
plt.title("Elbow Yöntemi (Inertia)")
plt.xlabel("K")
plt.ylabel("Inertia")

plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette, marker='o')
plt.title("Silhouette Skoru")
plt.xlabel("K")
plt.ylabel("Silhouette")

plt.tight_layout()
plt.show()
```

```python
optimal_k = 4  # örnek, grafikten belirle
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
df["Kume"] = kmeans.fit_predict(df)
```

```python
from sklearn.decomposition import PCA

# 2 boyutlu görselleştirme için PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df)

print("Toplam varyansın %", round(sum(pca.explained_variance_ratio_)*100, 2), "kadarı korunuyor.")
```

```python
# KMeans ile kümeleme (örneğin K=4)
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(df)

# Scatter plot çizimi
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=50)
plt.title("K-Means Kümeleme (PCA ile 2D)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Küme No")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wss = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(df)
    wss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, wss, marker='o')
plt.title("Elbow Yöntemi: WSS (Inertia)")
plt.xlabel("K (Küme Sayısı)")
plt.ylabel("WSS (Inertia)")
plt.grid(True)
plt.show()
```
